```python
import os
import time


#addition of numbers
def addition():   
    n = list(map(int, input("Enter all numbers seperated by space: ").split()))
    return sum(n)


#subtraction of numbers
def subtraction():
    n1 = float(input("Enter the first number: "))
    n2 = float(input("Enter the second number: "))
    s = n1 - n2
    return s


#multiplication of numbers
def multiplication():
    n = list(map(int, input("Enter all numbers seperated by space: ").split()))
    fnlrslt = 1
    for n in n:
        fnlrslt *= n
    return fnlrslt


#division of numbers
def division():
    n1 = float(input("Enter the first number: "))
    n2 = float(input("Enter the second number: "))
    d = n1 / n2
    return d


#average of numbers
def average():
    n = list(map(int, input("Enter all numbers seperated by space: ").split()))
    a = sum(n) / len(n)
    return a


#choosing the operation user want to conduct
c = 0
while c != "-1":
    print("Enter '+' for addition")
    print("Enter '-' for subtraction")
    print("Enter '*' for multiplication")
    print("Enter '/' for division")
    print("Enter '//' for average")
    print("Enter 'ext' to exit.\n")

    c = input("Your choice is: ")

    if c == "+":
        fnlrslt = addition()
        os.system("cls")
        print(f"The answer is {fnlrslt}")
        time.sleep(2)
        os.system("cls")

    elif c == "-":
        fnlrslt = subtraction()
        os.system("cls")
        print(f"The answer is {fnlrslt}")
        time.sleep(2)
        os.system("cls")

    elif c == "*":
        fnlrslt = multiplication()
        os.system("cls")
        print(f"The answer is {fnlrslt}")
        time.sleep(2)
        os.system("cls")

    elif c == "/":
        fnlrslt = division()
        os.system("cls")
        print(f"The answer is {fnlrslt}")
        time.sleep(2)
        os.system("cls")

    elif c == "//":
        fnlrslt = average()
        os.system("cls")
        print(f"The answer is {fnlrslt}")
        time.sleep(2)
        os.system("cls")

    elif c == "ext":
        os.system("cls")
        print("Thank you for using the calculator!")
        time.sleep(2)
        break
      

    else:
        os.system("cls")
        print("Sorry, invalid option!")
        time.sleep(2)
        os.system("cls")







    
```
